```
# 자판기 과제의 음료 판매 현황과 매출을 출력해봅시다

import numpy as np

음료 = np.array(['밀크티', '에너지드링크','콜라'])
가격 = np.array([700,800,600])
판매개수 = np.array([10,15,20])

매출 = 가격 * 판매개수
#print(매출)
```

```
# 음료별 판매개수 출력
plt.figure()
plt.title('판매개수')
plt.bar(음료, 판매개수)
plt.show()
print('총 판매개수는 {:,.0f}개입니다'.format(판매개수.sum()))

# 음료별 매출 출력
plt.figure()
plt.title('매출')
plt.bar(음료, 매출)
plt.show()
print('총매출은 {:,.0f}원입니다'.format(매출.sum()))
```
