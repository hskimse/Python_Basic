# loc
* loc는 인덱스를 가지고 특정 데이터 값을 선택한다. 

```python
import seaborn as sns # 그래픽 라이브러리이며, 데이터셋을 쉽게 불러올 수 있다
```

```python
mpg = sns.load_dataset('mpg') # 차량 무게 및 연비의 데이터입니다. 
```

```python
mpg
```

* loc[행의 조건, 열의 조건]을 통해서 필요한 부분만 선택가능함

```python
mpg.loc[0,'cylinders']
```

```python
mpg[:10]
```

```python
a= mpg.loc[:10, ['mpg', 'horsepower']]
a
```

```python
b = mpg.loc[:10, 'mpg': 'horsepower']
b
```

* 특정 컬럼만 가져오고 행은 전체행이라면 df[:,컬럼]으로 사용한다

```python
b2= mpg.loc[:,'mpg': 'horsepower']
b2
```

* 조건식을 행의 조건 안에 넣을 수 있다

```python
c = mpg.loc[mpg.cylinders==8, :]  # 앞에서 배운 query와 동일한 기능
c
```

```python
c = mpg.loc[mpg.cylinders==8]  # 앞에서 배운 query와 동일한 기능
c
```

```python
mpg.cylinders==8
```

* iloc는 인덱스의 값과 관계없이 순서대로 가져온다

```python
c.index = range(10,113)
c
```
