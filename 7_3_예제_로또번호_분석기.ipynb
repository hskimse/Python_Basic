# 로또 번호 분석기
* 로또 번호가 몇개가 얼마나 분포를 보이는 지 분석 합니다

```python
import random

# 랜덤 번호 생성을 함수로 만듬
def pick():    
    return random.sample(range(1,46),6)
```

```python
win = pick()
print('1등', win)
```

```python
# 여러개 로또 번호를 만들어 봅시다

n = 1000

results = []
for i in range(n):
    results.append(pick())
    
print(len(results))
```

```python
results
```

* 리스트에서 특정 범위 선택은 list[start:end]입니다
* range와 마찬가지로 end는 포함하지 않습니다
* start는 입력하지 않으면 0이 기본값입니다. 
* end는 입력하지 않으면 리스트의 마지막 값까지입니다

```python
results[:5]
```

```python
results[-5:]
```

```python
results[-20:-5]
```

* 정답과 확인해서 맞은 개수를 비교해봅시다

```python
pick_set = set(win) # 정답을 set으로 변경

n_corrects = []
for res in results:
    # 각각의 결과를 set으로 변경
    res_set = set(res)
    correct = pick_set & res_set
    #print(correct)
    n_correct = len(correct)
    n_corrects.append(n_correct)
n_corrects[:10]
    
    
```

* 리스트에 요소의 개수가 몇개인지 세는 것은 count()입니다. 

```python
n_corrects.count(1)
```

```python
# 0개부터 6개까지의 정답의 개수를 확인해봅시다

for i in range(7):
    print('{}개 맞춘 건수는 {}건입니다. '.format(i, n_corrects.count(i)))
```

# 코드 정리

```python
import random

# 랜덤 번호 생성을 함수로 만듬
def pick():
    return random.sample(range(1,46),6)

win = pick()
print('1등', win)

# 여러개 로또 번호를 만들어 봅시다
n = 1000000

results = []
for i in range(n):
    results.append(pick())
    
pick_set = set(win) # 정답을 set으로 변경

n_corrects = []
for res in results:
    # 각각의 결과를 set으로 변경
    res_set = set(res)
    correct = pick_set & res_set
    n_correct = len(correct)
    n_corrects.append(n_correct)

# 0개부터 6개까지의 정답의 개수를 확인해봅시다

for i in range(7):
    print('{}개 맞춘 건수는 {}건입니다. '.format(i, n_corrects.count(i)))

for i in range(7):
    print('{}개 맞출 확률은 {}건입니다. '.format(i, n_corrects.count(i)/n))
```
