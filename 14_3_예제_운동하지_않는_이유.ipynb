```
import pandas as pd

data = pd.read_csv('/g/My Drive/Python_202002/report.txt', sep = '\t') #utf-8로 인코딩 되어 있음 
print(data.columns)
```

```
data.head()
```

```
data
```

```
# 불필요한 데이터를 삭제해봅시다 
# 기간 컬럼을 삭제해보겠습니다 .
df = data.drop(columns = '기간')

# 지역구별로 분류된 행들을 삭제해보겠습니다. 
df.drop(index = range(27, 52), inplace=True)
```

```
df
```

```
# 대분류가 성별인 데이터만 가져옵니다. 
df = df.query('대분류=="성별"')
#df = df.loc[df.대분류=='성별'] #loc를 활용해서 선택
```

```
df
```

```
df = df.drop(columns = '대분류')
df.set_index('분류', inplace=True)
df_T = df.T # 행과 열을 바꿈
```

```
df_T
```

```
df_T.dtypes
```

```
df_T.남자 = df_T.남자.astype('float') #문자를 숫자로 변경
df_T.여자 = df_T.여자.astype('float')
```

```
# 한글깨짐 방지
import matplotlib as mpl
import matplotlib.pyplot as plt
 
%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
import matplotlib.font_manager as fm
fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()
```

```
plt.figure()
df_T['남자'].plot.pie(autopct = '%1.1f%%')
plt.title('남자')
plt.show()

plt.figure()
df_T['여자'].plot.pie(autopct = '%1.1f%%')
plt.title('여자')
plt.show()
```
