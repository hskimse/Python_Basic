```
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/My Drive/Python_202002/forbes_celebrity_100.csv')
data.rename(columns={'Pay (USD millions)':'Pay_USD'}, inplace=True) 

```

```
data
```

```
data.dtypes
```

```
# 2019년 actor의 가장 소득 많은 사람을 뽑아보자
df_actor = data.query('Category=="Actors" & Year ==2019')
```

```
df_actor
```

```
top_alltime = data.sort_values('Pay_USD', ascending = False).head(5)
```

```
top_alltime
```

```
df_actor.plot.bar(x='Name', y='Pay_USD')
```

```
# 매년 가장 소득 많은 3명을 뽑아보자
df_group = data.groupby('Year')

```

```
df_group.head()
```

```
df_group.get_group(2015)
```

```
df_group.groups.keys()
```

```
df_group = df_group.apply(lambda df: df.sort_values('Pay_USD', ascending = False).head(3))
```

```
df_group
```

```
df = df_group.reset_index(drop=True) # multi index를 없앰
```

```
df
```

```
# pivot_table을 통해 차트를 그리기 위한 데이터 모양으로 변경
df2 = pd.pivot_table(data, values='Pay_USD', index=['Year'],
                    columns=['Category'], aggfunc = 'sum')
```
