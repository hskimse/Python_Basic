# 그래프 - Matplotlib
* 파이썬에는 다양한 그래프 라이브러리가 있으며, matplotlib 가 가장 대표적인 그래프 라이브러리입니다
* https://matplotlib.org/ 에 다양한 그래프에 대한 예제가 있습니다.
* 우리가 사용할 그래프들은 matplotlib.pyplot에 있습니다

```python
# 리스트를 차트로 그려봅시다

import matplotlib.pyplot 
매출 = [100,150,200,180,400,1000]
matplotlib.pyplot.plot(매출)
```

* import 라이브러리명 as 명칭 으로 긴 이름을 짧게 바꿔 사용합니다

```python
import matplotlib.pyplot as plt

매출 = [100,150,200,180,400,1000]
plt.plot(매출)
plt.show()
```

```python
매출 = [100,150,200,180,400,1000]
매출2 = [150,200,180,400,1000,900]
plt.plot(매출)
plt.show()  # 이 줄을 삭제/주석 처리하면 한개의 그래프에 그려진다. 
plt.plot(매출2)
plt.show()
```

* x축을 추가해봅시다

```python
매출 = [100,150,200,180,400,1000]
월 =[3,4,5,6,7,8]
plt.plot(월,매출)
plt.show()
```

# bar plot
* 위의 그래프를 막대그래프로 그려봅시다
* bar()를 사용합니다. 

```python
plt.bar(월, 매출)
plt.show()
```

* 한글 사용 시 아래처럼 깨지게 됩니다

* 제목: plt.title()
* x축 제목: plt.xlabel()
* y축 제목: plt.ylabel()
* 눈금 보이기: plt.grid()

```python
plt.title('월별 매출 동향')
plt.ylabel('매출액(만원)')
plt.xlabel('월')
plt.grid()
plt.bar(월, 매출)
plt.show()
```

* 색상 변경하기
* color를 입력하면 변경됩니다. 
* 색상 입력은 https://matplotlib.org/examples/color/named_colors.html 를 참고하면됩니다

```python
plt.bar(월, 매출, color = ['gray','gray','gray','gray','tomato','r'])
plt.show()
```

# histogram
* 빈도수 확인을 위한 그래프를 출력
* plt.hist() 사용

```python
# 주사위를 무작위로 던져서 나온 빈도수 확인

import random
samples = random.choices(list(range(1,7)), k=1000)
print(samples)
#plt.hist(samples)
plt.hist(samples, bins = range(1,8), align='left')
#plt.hist(samples, bins = range(1,8), align = 'left')
plt.show()
```

# pie chart
* plt.pie() 사용
* 값과 항목명을 넣어주면 된다. 

```python
득표수 = [100,80,300,10]
후보 = ['1번','2번', '3번', '4번']
plt.pie(득표수,labels= 후보, autopct = '%.1f%%')
#plt.pie(득표수,labels= 후보)
plt.show()
```
