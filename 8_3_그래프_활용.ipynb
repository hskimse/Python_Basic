```
import matplotlib.pyplot as plt
import numpy as np

# 사용자 입력
x_start = -1
x_end = 100
func_y = lambda x: x**2.5 - x**2 +10*x +4

# def func_y(x):
#     return x**3 + x**2 -100*x +4

# 함수를 출력하는 함수를 만들어봄
def plot_func(start, end, func): # 함수를 입력으로 받을 수 있음
    x = np.linspace(start, end, num=100)
    y = func(x)  
    plt.plot(x,y)      
    plt.show()
    
plot_func(x_start, x_end, func_y)
```

```
# 2개의 함수를 한개의 차트에 그리기
# 사용자 입력
x_start = -10
x_end = 10

func_y1 = lambda x: x**2 +10
func_y2 = lambda x: x*2 + 10


def plot_twofunc(start, end, func1, func2):
    x = np.linspace(start, end, num=100)
    y1 = func1(x)
    y2 = func2(x)
    plt.plot(x,y1)   
    plt.plot(x,y2)        
    plt.show()


plot_twofunc(x_start, x_end, func_y1, func_y2)
```
